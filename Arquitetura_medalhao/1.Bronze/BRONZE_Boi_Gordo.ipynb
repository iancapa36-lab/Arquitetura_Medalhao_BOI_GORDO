```python
# Conectando o DataFrame com a tabela do cepea

df = spark.table("workspace.bronze_economia.boi_gordo_cepea")
```

```python
df.display()
```

```python
# Essa funcao que eu fiz utilizando o current_timestamp() para adicionar uma coluna com a data e hora atual com a data da coleta dos dados 

from pyspark.sql.functions import current_timestamp
df = df.withColumn("data_coleta", current_timestamp())
df.display()
```

```python
# Salvar essa tabela no DataFrame(df) utilizando o df.write formato delta mode overwrit salvando a tabela no catalogo bronze nomeando a tabela como boi_gordo, pois ja t

df.write.format("delta").mode("overwrite").saveAsTable(f"{CATALOG}.{BRONZE}.boi_gordo")
```
